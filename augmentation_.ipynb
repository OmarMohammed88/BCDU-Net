```
import imageio
import imgaug as ia
from imgaug import augmenters as iaa
import os
import numpy as np

from imgaug.augmentables.heatmaps import HeatmapsOnImage
```

```
seq = iaa.Sequential([
    iaa.Affine(rotate=25),
    iaa.Fliplr(0.5),
    iaa.Sharpen(alpha=0.5,lightness=(0.75, 2.0))
])
```

```
sharp = iaa.Sharpen(alpha=0.5,lightness=(0.75, 2.0))
```

```
crop1= iaa.Crop(percent=(0, 0.2))
```

```
crop0= iaa.Crop(percent=(0, 0.4))
```

```
image_list1=os.listdir('/content/growth_stage_1')
image_list1.remove('output')
for i in image_list1:
  image = imageio.imread("/content/growth_stage_1/"+i)
  image_sharp=sharp(image=image)
  images_seq = seq(image=image)
  image_crop1=crop1(image=image)
  image_crop0=crop0(image=image)
  imageio.imwrite('/content/growth_stage_1/output/'+'sharp'+i, image_sharp)
  imageio.imwrite('/content/growth_stage_1/output/'+'crop1'+i, image_crop1)
  imageio.imwrite('/content/growth_stage_1/output/'+'crop0'+i, image_crop0)
  imageio.imwrite('/content/growth_stage_1/output/'+'seq'+i, images_seq)
```
